### âœ… 1. â€œæŒ‰ç…§è‡ªå·±éœ€æ±‚æ­å»ºæ¨¡å‹â€ 
ä½¿ç”¨ tf.keras.Model å­ç±»åŒ–ï¼Œä½ å¯ä»¥ï¼š\
1ã€è‡ªç”±å®šä¹‰ __init__ ä¸­çš„å±‚ï¼ˆå¦‚ Denseã€Convã€LSTMã€è‡ªå®šä¹‰å±‚ç­‰ï¼‰\
2ã€çµæ´»ç»„åˆå„ç§å±‚ï¼Œæ„å»ºéå¸¸è§„ç»“æ„çš„æ¨¡å‹ï¼Œæ¯”å¦‚ï¼š\
å¤šè¾“å…¥ / å¤šè¾“å‡ºæ¨¡å‹\

GAN çš„ç”Ÿæˆå™¨æˆ–åˆ¤åˆ«å™¨\
å¼ºåŒ–å­¦ä¹ ä¸­çš„ç­–ç•¥ç½‘ç»œç­‰\
### âœ… ä¸¾ä¾‹ï¼š

```python
class MyModel(tf.keras.Model):
    def __init__(self):
        super().__init__()   # è‡ªåŠ¨ç»§æ‰¿çˆ¶ç±»çš„æ„é€ æ–¹æ³•
        self.dense1 = tf.keras.layers.Dense(64, activation='relu')   
        self.dense2 = tf.keras.layers.Dense(32)
        self.skip_connect = tf.keras.layers.Dense(32)  # å¯ä»¥åšæ®‹å·®è¿æ¥
        self.output_layer = tf.keras.layers.Dense(10, activation='softmax')  # è¾“å‡ºå±‚ å¤šåˆ†ç±»è¾“å‡º 

è¿™æ¯” Sequential æ¨¡å‹çµæ´»å¾—å¤šã€‚
```

### 2. â€œè‡ªå·±æ§åˆ¶æ•°æ®çš„æµå‘â€ âœ… æ­£ç¡®
åœ¨ call(self, inputs) æ–¹æ³•ä¸­ï¼Œä½ æ‰‹åŠ¨ç¼–å†™å‰å‘ä¼ æ’­é€»è¾‘ï¼Œå†³å®šæ•°æ®å¦‚ä½•æµåŠ¨ã€‚

```python
def call(self, x):
    h1 = self.dense1(x)
    h2 = self.dense2(h1)
    skip = self.skip_connect(x)  # ç›´æ¥ä»è¾“å…¥å¼•å‡ºä¸€æ”¯è·¯
    h2 = h2 + skip  # å®ç°æ®‹å·®è¿æ¥
    return self.output_layer(h2)
```

ä½ å¯ä»¥ï¼š\
å†™ if æ¡ä»¶åˆ¤æ–­ï¼ˆæ¯”å¦‚è®­ç»ƒ/æ¨ç†ä¸åŒè¡Œä¸ºï¼‰\
å†™ for å¾ªç¯ï¼ˆå¦‚ RNN å±•å¼€ï¼‰\
å®ç°å¤æ‚çš„è¿æ¥æ–¹å¼ï¼ˆå¦‚è·³è·ƒè¿æ¥ã€å¤šè·¯å¾„ã€æ³¨æ„åŠ›æƒé‡è®¡ç®—ç­‰ï¼‰\
âœ… è¿™æ˜¯ Sequential æ¨¡å‹åšä¸åˆ°çš„ã€‚

### âš ï¸ ä½†è¦æ³¨æ„ï¼šä¸æ˜¯å”¯ä¸€çš„â€œè‡ªå®šä¹‰â€æ–¹å¼
Keras æä¾›äº†ä¸‰ç§æ„å»ºæ¨¡å‹çš„æ–¹å¼ï¼Œtf.keras.Model å­ç±»åŒ–åªæ˜¯å…¶ä¸­ä¹‹ä¸€ï¼š\

| å‡½æ•°å¼ APIï¼ˆFunctional APIï¼‰ | ä¸­     | ä¸­     | æœ‰åˆ†æ”¯ã€å…±äº«å±‚ã€å¤šè¾“å…¥è¾“å‡ºçš„é™æ€å›¾æ¨¡å‹ |
| Model å­ç±»åŒ–       | é«˜     | ä½     | åŠ¨æ€é€»è¾‘ã€å¤æ‚æ§åˆ¶æµã€ç ”ç©¶å‹æ¨¡å‹ |

âœ… æ‰€ä»¥ä½ é€‰æ‹© Model å­ç±»åŒ–ï¼Œå°±æ˜¯ä¸ºäº†æœ€å¤§è‡ªç”±åº¦ã€‚

### âœ… ä¸¾ä¸ªä¾‹å­è¯´æ˜â€œæ§åˆ¶æ•°æ®æµå‘â€çš„å¼ºå¤§

```python
def call(self, x, training=None):
    if training:
        noise = tf.random.normal(tf.shape(x)) * 0.1
        x = x + noise  # åªåœ¨è®­ç»ƒæ—¶åŠ å™ªå£°ï¼ˆå¦‚å»å™ªè‡ªç¼–ç å™¨ï¼‰
    return self.decoder(self.encoder(x))
```

è¿™ç§è®­ç»ƒ/æ¨ç†è¡Œä¸ºä¸åŒçš„é€»è¾‘ï¼Œåªèƒ½é€šè¿‡å­ç±»åŒ–å®ç°ã€‚

## âœ… æ€»ç»“

ä½¿ç”¨ tf.keras.Model å­ç±»åŒ–ï¼Œå°±æ˜¯ä¸ºäº†è®©å¼€å‘è€…åƒå†™æ™®é€š   å‡½æ•°ä¸€æ ·ï¼Œè‡ªç”±å®šä¹‰æ¨¡å‹ç»“æ„å’Œæ•°æ®æµåŠ¨é€»è¾‘ï¼Œç‰¹åˆ«é€‚åˆç ”ç©¶ã€å®éªŒã€å¤æ‚ç½‘ç»œç»“æ„ã€‚

ğŸ“Œ å»ºè®®ï¼šåœ¨èƒ½ç”¨ å‡½æ•°å¼ API è§£å†³æ—¶ä¼˜å…ˆç”¨å®ƒï¼ˆæ›´ç®€æ´ã€å¯è¯»æ€§å¥½ï¼‰ï¼›åªæœ‰åœ¨éœ€è¦åŠ¨æ€æ§åˆ¶æµæ—¶ï¼Œæ‰ä½¿ç”¨ Model å­ç±»åŒ–ã€‚

# ä¸‰ã€å®Œæ•´ç¤ºä¾‹ä»£ç ï¼ˆæ‰‹æŠŠæ‰‹ï¼‰

```python
import tensorflow as tf

# 1. å®šä¹‰æ¨¡å‹ç±»
class MyModel(tf.keras.Model):
    def __init__(self):
        super().__init__()
        self.dense1 = tf.keras.layers.Dense(64, activation='relu')
        self.dense2 = tf.keras.layers.Dense(10, activation='softmax')

    def call(self, inputs):   # æ§åˆ¶æ•°æ®åˆ†å‘
        x = self.dense1(inputs)
        return self.dense2(x)

# 2. å®ä¾‹åŒ–æ¨¡å‹
model = MyModel()

# 3. ç¼–è¯‘æ¨¡å‹
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# å‡æ•°æ®ç¤ºèŒƒ
import numpy as np
train_x = np.random.random((1000, 20))
train_y = np.random.randint(0, 10, (1000,))

test_x = np.random.random((200, 20))
test_y = np.random.randint(0, 10, (200,))

# 4. è®­ç»ƒæ¨¡å‹
model.fit(train_x, train_y, epochs=5, batch_size=32)

# 5. è¯„ä¼°æ¨¡å‹
model.evaluate(test_x, test_y)

# 6. é¢„æµ‹
sample_data = np.random.random((5, 20))
predictions = model.predict(sample_data)
print(predictions)
```

# äº”ã€call æ–¹æ³•é¢å¤–å‚æ•°è¯´æ˜
call(self, inputs, training=False) ä¸­ï¼š

inputsï¼šè¾“å…¥å¼ é‡ã€‚\
trainingï¼šå¸ƒå°”å€¼ï¼Œè¡¨ç¤ºå½“å‰æ˜¯è®­ç»ƒè¿˜æ˜¯æ¨ç†é˜¶æ®µï¼ˆæœ‰äº›å±‚å¦‚ Dropout, BatchNorm éœ€è¦åŒºåˆ†ï¼‰ã€‚\
ä½ å¯ä»¥æ ¹æ® training å‚æ•°è®¾è®¡è®­ç»ƒå’Œæ¨ç†ä¸åŒé€»è¾‘ï¼š

```python
def call(self, inputs, training=False):
    x = self.dense1(inputs)
    if training:
        x = tf.nn.dropout(x, rate=0.5)
    return self.dense2(x)
```

## 2ã€å‡½æ•°å¼ API åŸºæœ¬æ­¥éª¤å’Œç¤ºä¾‹

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

# 1. å®šä¹‰è¾“å…¥å¼ é‡ï¼ˆå¿…é¡»ï¼‰
inputs = tf.keras.Input(shape=(20,))  # è¾“å…¥ç‰¹å¾ç»´åº¦æ˜¯20

# 2. å®šä¹‰ç½‘ç»œå±‚å¹¶è¿æ¥ï¼ˆåƒæ­ç§¯æœ¨ï¼Œå±‚å’Œå¼ é‡äº’ç›¸è°ƒç”¨ï¼‰
x = layers.Dense(64, activation='relu')(inputs)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(10, activation='softmax')(x)

# 3. åˆ›å»ºæ¨¡å‹ï¼ŒæŒ‡å®šè¾“å…¥å’Œè¾“å‡º
# name è¿™ä¸ªå‚æ•°ï¼Œæ˜¯ç»™ä½ åˆ›å»ºçš„æ¨¡å‹æŒ‡å®šä¸€ä¸ªåå­—ï¼ˆnameï¼‰ï¼Œæ–¹ä¾¿ä½ åœ¨åç»­ä½¿ç”¨æ—¶è¾¨è¯†å’Œç®¡ç†ã€‚
model = Model(inputs=inputs, outputs=outputs, name='functional_model')

# 4. ç¼–è¯‘æ¨¡å‹
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 5. è®­ç»ƒ
# å‡æ•°æ®ç¤ºä¾‹
import numpy as np
train_x = np.random.random((1000, 20))
train_y = np.random.randint(0, 10, (1000,))

model.fit(train_x, train_y, epochs=5, batch_size=32)


# name å‚æ•°çš„è®²è§£
model = Model(inputs=inputs, outputs=outputs, name='functional_model')
print(model.name) 
# è¾“å‡ºï¼šfunctional_model

å¦‚æœä¸æŒ‡å®šï¼Œç³»ç»Ÿé»˜è®¤ä¼šç»™ä½ ä¸€ä¸ªåå­—ï¼Œå¦‚ model_1ã€‚
```

## 3ã€é‡è¦ç‚¹è¯´æ˜

| å±‚çš„è°ƒç”¨                     | å±‚æœ¬èº«æ˜¯å¯è°ƒç”¨çš„å‡½æ•°ï¼Œè¾“å…¥å¼ é‡è¿›å»ï¼Œè¾“å‡ºæ–°çš„å¼ é‡ã€‚     |
| `Model(inputs, outputs)` | åˆ›å»ºæ¨¡å‹å®ä¾‹ï¼Œå¿…é¡»æ˜ç¡®è¾“å…¥å’Œè¾“å‡ºã€‚             |
| çµæ´»è¿æ¥                     | å±‚ä¸å±‚ä¹‹é—´å¯ä»¥ä»»æ„è¿æ¥ï¼Œæ”¯æŒå¤šè¾“å…¥å¤šè¾“å‡ºã€æ®‹å·®ç­‰å¤æ‚ç»“æ„ã€‚ |
## å¤šè¾“å…¥å¤šè¾“å‡ºç¤ºä¾‹ 


```python
# å¤šè¾“å…¥
input_a = tf.keras.Input(shape=(32,))
input_b = tf.keras.Input(shape=(64,))

x1 = layers.Dense(16, activation='relu')(input_a)
x2 = layers.Dense(16, activation='relu')(input_b)

# æŠŠå¤šä¸ªå¼ é‡ï¼ˆTensorï¼‰åœ¨æŒ‡å®šçš„ç»´åº¦ä¸Šæ‹¼æ¥ï¼ˆConcatenateï¼‰æˆä¸€ä¸ªå¼ é‡ã€‚
# å¯ä»¥æŠŠå®ƒæƒ³è±¡æˆ NumPy é‡Œçš„ np.concatenate()ï¼Œåªæ˜¯å®ƒæ˜¯ Keras å±‚ç‰ˆæœ¬ï¼Œå¯ä»¥ç›´æ¥åœ¨æ¨¡å‹é‡Œç”¨ï¼Œå¹¶ä¸”æ˜¯å¯è®­ç»ƒè®¡ç®—å›¾çš„ä¸€éƒ¨åˆ†ã€‚
concat = layers.concatenate([x1, x2])
output = layers.Dense(1, activation='sigmoid')(concat)

model = Model(inputs=[input_a, input_b], outputs=output)

model.compile(optimizer='adam', loss='binary_crossentropy')
```
