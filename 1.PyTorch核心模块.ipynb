### (2) éæ ‡é‡å¼ é‡

```python
x = torch.tensor([[1., 2.], [3., 4.]], requires_grad=True)
y = x * 2           # y.shape = (2,2)ï¼Œä¸æ˜¯æ ‡é‡

grad_outputs = torch.ones_like(y)   # æ¯ä¸ªå…ƒç´ æƒé‡=1
torch.autograd.backward(y, grad_tensors=grad_outputs)
print(x.grad)
```
