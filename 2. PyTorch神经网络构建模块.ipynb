nn.Module å°±åƒ ä¹é«˜ç§¯æœ¨çš„åŸºç±» â€”â€” ä½ å®šä¹‰çš„æ¯ä¸ªç§¯æœ¨ï¼ˆå±‚/ç½‘ç»œï¼‰éƒ½è¦ç»§æ‰¿å®ƒï¼ŒæŠŠå±‚åœ¨ __init__ é‡Œæ­å¥½ï¼ŒæŠŠè®¡ç®—è¿‡ç¨‹å†™åœ¨ forward é‡Œï¼Œç„¶å PyTorch ä¼šå¸®ä½ è‡ªåŠ¨å¤„ç†å‚æ•°ç®¡ç†ã€æ¢¯åº¦è®¡ç®—ã€ä¿å­˜åŠ è½½ç­‰ã€‚

# ğŸ”¹å®Œæ•´ç¤ºä¾‹ï¼šMNIST åˆ†ç±»

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# 1. æ•°æ®é¢„å¤„ç†
transform = transforms.Compose([
    transforms.ToTensor(),                # è½¬æ¢ä¸º Tensor
    transforms.Normalize((0.5,), (0.5,))  # æ ‡å‡†åŒ–åˆ° [-1,1]
])


train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset  = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# ç”¨ DataLoaderï¼Œå®ƒè´Ÿè´£é«˜æ•ˆè¿­ä»£å·²åˆ’åˆ†å¥½çš„æ•°æ®ï¼Œä¸æ¶‰åŠåˆ’åˆ†é€»è¾‘ï¼Œä¸ºæ•°æ®é›†æä¾›æ‰¹é‡åŠ è½½å’Œè¿­ä»£çš„åŠŸèƒ½
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader  = DataLoader(test_dataset, batch_size=1000, shuffle=False)


# 2. å®šä¹‰ç½‘ç»œç»“æ„
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(28*28, 256)   # è¾“å…¥å±‚ â†’ éšè—å±‚1
        self.fc2 = nn.Linear(256, 128)     # éšè—å±‚1 â†’ éšè—å±‚2
        self.fc3 = nn.Linear(128, 10)      # éšè—å±‚2 â†’ è¾“å‡ºå±‚

        self.relu = nn.ReLU()              # æ¿€æ´»å‡½æ•°
        self.dropout = nn.Dropout(0.5)     # Dropout é˜²æ­¢è¿‡æ‹Ÿåˆ

    def forward(self, x):
        x = x.view(-1, 28*28)              # å±•å¹³
        x = self.relu(self.fc1(x))         # å…¨è¿æ¥ + ReLU
        x = self.dropout(x)                # Dropout
        x = self.relu(self.fc2(x))         # å…¨è¿æ¥ + ReLU
        x = self.fc3(x)                    # è¾“å‡ºå±‚ï¼ˆæœªåšSoftmaxï¼‰
        return x                           # äº¤å‰ç†µå†…éƒ¨ä¼šåšLogSoftmax


# 3. åˆå§‹åŒ–æ¨¡å‹ã€æŸå¤±å‡½æ•°ã€ä¼˜åŒ–å™¨
model = Net()
criterion = nn.CrossEntropyLoss()          # åˆ†ç±»ä»»åŠ¡å¸¸ç”¨
optimizer = optim.Adam(model.parameters(), lr=0.001)


# 4. è®­ç»ƒå‡½æ•°
def train(model, loader, optimizer, criterion, device="cpu"):
    model.train()
    total_loss, correct = 0, 0
    for data, target in loader:
        data, target = data.to(device), target.to(device)

        optimizer.zero_grad()
        output = model(data)

        loss = criterion(output, target)   # è®¡ç®—æŸå¤±
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        pred = output.argmax(dim=1)        # é¢„æµ‹ç±»åˆ«
        correct += (pred == target).sum().item()

    acc = correct / len(loader.dataset)
    return total_loss / len(loader), acc


# 5. æµ‹è¯•å‡½æ•°
def evaluate(model, loader, criterion, device="cpu"):
    model.eval()
    total_loss, correct = 0, 0
    with torch.no_grad():
        for data, target in loader:
            data, target = data.to(device), target.to(device)
            output = model(data)

            loss = criterion(output, target)
            total_loss += loss.item()
            pred = output.argmax(dim=1)
            correct += (pred == target).sum().item()

    acc = correct / len(loader.dataset)
    return total_loss / len(loader), acc


# 6. è®­ç»ƒ + æµ‹è¯•å¾ªç¯
for epoch in range(1, 6):  # è®­ç»ƒ 5 è½®
    train_loss, train_acc = train(model, train_loader, optimizer, criterion)
    test_loss, test_acc = evaluate(model, test_loader, criterion)
    print(f"Epoch {epoch}: Train Loss={train_loss:.4f}, Train Acc={train_acc:.4f}, "
          f"Test Loss={test_loss:.4f}, Test Acc={test_acc:.4f}")
```
